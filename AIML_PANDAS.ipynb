```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/content/vgsalesGlobale.csv")
```

```python
df.head(15)
```

```python
sliced_df = df[['Name', 'Global_Sales']].head(5)
print("\nSliced Dataset:")
print(sliced_df)
```

```python
counts = df.count()
print("\nCount of Non-Null Values in Each Column:")
print(counts)
```

```python
df_no_null = df.dropna()
print("\nDataset after Removing Null Values:")
print(df_no_null)
```

```python
df_filled = df.copy()
df_filled['Year'].fillna(df_filled['Year'].median(), inplace=True)
print("\nDataset after Filling Null Values in 'Year' Column:")
print(df_filled)
```

```python
sales_gt_30m = df[df['Global_Sales'] > 30]
print("\nRows with Global Sales > 30 million:")
print(sales_gt_30m)
```

```python
recent_games = df.loc[df['Year'] > 2010, ['Name', 'Global_Sales']]
print("\nRecent Games (Name and Global Sales) released after 2010:")
print(recent_games)
```

```python
selected_data = df.iloc[2:6, 1:5]
print("\nSelected Rows (2 to 5) and Columns (1 to 4):")
print(selected_data)
```

```python
wii_games = df[df['Platform'] == 'Wii']
print("\nWii Games:")
print(wii_games)
```

```python
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.show()
```

```python
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_publishers.plot(kind='bar', color='skyblue')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.title('Top 10 Publishers by Global Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Global_Sales'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.title('Distribution of Global Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Global_Sales'], color='orange', marker='o')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.title('Global Sales vs Year')
plt.show()
```
